# Exercise 1

Our original assumption is that $\mathbf{y}$ is given by 

\begin{equation}
    \mathbf{y} = f(x) + \epsilon,
\end{equation}
where $f(x)$ is a continuous function and $\epsilon$ is Gaussian distributed. Now, we approximate this function $f(x)$ by $f(x) \approx \tilde{\mathbf{y}} = \mathbf{X} \mathbf{\beta}$, such that 

\begin{equation}
    \mathbf{y} \approx \mathbf{X} \mathbf{\beta} + \epsilon,
\end{equation}
or
\begin{equation}
    y_i \approx \sum_{j} X_{ij} \beta_{j} + \epsilon_i =\mathbf{X}_i \cdot \mathbf{\beta} + \epsilon_i.
\end{equation}

Now, if we take the expectation value of this, we get

\begin{align*}
    \mathbb{E} \left[y_i \right] &\approx \mathbb{E} \left[ \mathbf{X}_i \cdot \mathbf{\beta}\right] + \mathbb{E} \left[\epsilon_i\right] \\
    &= \mathbf{X}_i \cdot \mathbf{\beta},
\end{align*}
because $\mathbf{X}$ and $\mathbf{\beta}$ are assumed to be non-stochastic variables (enforced by design), and $\mathbf{\epsilon} = 0$ by the properties of a normal distribution with zero mean.

For the variance, we have that

\begin{align*}
    Var \left[y_i\right] &= \mathbb{E} \left[ \left( y_i - \mathbb{E} \left[y_i \right] \right)^2\right] \\
    &= \mathbb{E} \left[y_i^2 - 2y_i\mathbb{E}\left[y_i\right] + \mathbb{E}\left[y_i\right]^2\right] \\
    &= \mathbb{E} \left[\left(\mathbf{X}_i \cdot \mathbf{\beta} + \epsilon_i\right)^2 - 2\left(\mathbf{X}_i \cdot \mathbf{\beta} + \epsilon_i\right)\mathbf{X}_i \cdot \mathbf{\beta} + \left(\mathbf{X}_i \cdot \mathbf{\beta}\right)^2\right] \\
    &= \mathbb{E} \left[\left(\mathbf{X}_i \cdot \mathbf{\beta}\right)^2 + 2\mathbf{X}_i \cdot \mathbf{\beta}\epsilon + \epsilon^2 - 2\left(\mathbf{X}_i \cdot \mathbf{\beta}\right)^2 - 2\mathbf{X}_i \cdot \beta \epsilon + \left(\mathbf{X}_i \cdot \mathbf{\beta}\right)^2\right] \\
    &= \mathbb{E} \left[\epsilon^2\right] \\
    &= \sigma^2 \\
\end{align*}

For the expectation value of $\hat{\mathbf{\beta}}$, we note that since $\mathbf{X}$ is non-stochastic, the inverse $\left( \mathbf{X}^T \mathbf{X} \right)^{-1}$ and product with $\mathbf{X}^T$ is also non-stochastic, so
\begin{align*}
    \mathbb{E} \left[ \hat{\mathbf{\beta}} \right] &= \mathbb{E} \left[ \left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \mathbf{y} \right] \\
    &= \left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \mathbb{E} \left[ \mathbf{y} \right] \\
    &= \left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \left( \mathbb{E} \left[ \mathbf{X}\beta\right] + \mathbb{E} \left[\epsilon\right] \right) \\
    &= \left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \mathbf{X}\beta \\
    &= \left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \tilde{\mathbf{y}} \\
    &= \beta
\end{align*}


Finally, the variance of $\hat{\mathbf{\beta}}$, given by

\begin{align*}
    Var \left[\hat{\mathbf{\beta}} \right] &= \mathbb{E} \left[ \left( \hat{\mathbf{\beta}} - \mathbb{E} \left[\hat{\mathbf{\beta}}  \right] \right)^2\right] \\
    &= \mathbb{E} \left[\hat{\mathbf{\beta}}^2 - 2\hat{\mathbf{\beta}}\beta +\beta^2 \right] \\
\end{align*}

Now let's do some intermediate, simplifying calculations:

\begin{align*}
    \hat{\mathbf{\beta}} &= \left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \mathbf{y} \\
    &= \left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \left(\mathbf{X} \beta + \epsilon\right)  \\
    &= \left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \tilde{\mathbf{y}} + \left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \epsilon  \\
    &= \beta + \left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \epsilon
\end{align*}

\begin{align*}
    \hat{\mathbf{\beta}}^2 &= \beta^2 +2\beta\left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \epsilon + \left[\left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \epsilon\right]^2
\end{align*}


\begin{align*}
    -2\hat{\mathbf{\beta}} \beta &= -2\beta^2 - 2\beta\left(\mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \epsilon
\end{align*}

Putting it all together we get

\begin{align*}
    Var \left[\hat{\mathbf{\beta}} \right] &= \mathbb{E} \left[ \beta^2 +2\beta\left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \epsilon + \left[\left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \epsilon\right]^2 -2\beta^2 - 2\beta\left(\mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \epsilon + \beta^2 \right] \\
    &= \mathbb{E} \left[ \left(\left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^T \epsilon\right)^2 \right] \\
    &= \left(\left( \mathbf{X}^T \mathbf{X} \right)^{-1}\right)^2 \left( \mathbf{X}^T \mathbf{X}\right) \mathbb{E} \left[ \epsilon^2\right] \\
    &= \left( \mathbf{X}^T \mathbf{X} \right)^{-1} \sigma^2 \\
    &= \sigma^2 \left( \mathbf{X}^T \mathbf{X} \right)^{-1}
\end{align*}
assuming that all compounds of $\mathbf{X}$ are non-stochastic and the product $\left(\mathbf{X}^T\right)^2 = \mathbf{X}^T \mathbf{X}$ is properly defined (the notation might be a bit sloppy shorthand, but simplifications like $\left(\left( \mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X^T} \right)^2 = \left( \mathbf{X}^T \mathbf{X} \right)^{-1}$ using $\left(\mathbf{A}_1 \mathbf{A}_2 \cdots \mathbf{A}_n\right)^T = \mathbf{A}^T_n \cdots  \mathbf{A}^T_2  \mathbf{A}^T_1$ and $\left(\mathbf{A}_1 \mathbf{A}_2 \cdots \mathbf{A}_n\right)^{-1} = \mathbf{A}^{-1}_n \cdots  \mathbf{A}^{-1}_2  \mathbf{A}^{-1}_1$ are a recurring theme, which corresponds to "normal algebra").

Before we start, we note that the same arguments apply to non-stochastic variables, now including ridge parameter $\lambda$ and the identity matrix.

\begin{align*}
    \mathbb{E} \left[\hat{\beta}^{\rm Ridge} \right] &= \mathbb{E} \left[ \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T \mathbf{y} \right] \\
    &=  \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T \mathbb{E} \left[\mathbf{y} \right] \\
    &= \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T \left( \mathbb{E} \left[ \mathbf{X}\beta\right] + \mathbb{E} \left[\epsilon\right] \right) \\
    &= \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T \mathbf{X} \mathbb{E} \left[ \beta\right] \\
    &= \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T \mathbf{X} \beta \\
\end{align*}


For the variance,

\begin{align*}
    Var \left[\hat{\mathbf{\beta}}_{\rm Ridge} \right] &= \mathbb{E} \left[ \left( \hat{\mathbf{\beta}}_{\rm Ridge} - \mathbb{E} \left[\hat{\mathbf{\beta}}_{\rm Ridge}  \right] \right)^2\right] \\
    &= \mathbb{E} \left[ \left( \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T \mathbf{y}  - \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T \mathbf{X} \beta \right)^2 \right] \\
    &= \mathbb{E} \left[ \left( \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T\left(\mathbf{y} - \mathbf{X}\beta \right)\right)^2\right] \\
    &= \mathbb{E} \left[ \left( \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T \epsilon \right)^2\right] \\
    &= \mathbb{E} \left[ \left( \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T \epsilon \right)\left( \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T \epsilon \right)^T\right] \\
    &= \mathbb{E} \left[  \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T \epsilon \epsilon^T \mathbf{X} \left\{ \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \right\}^T \right] \\
    &= \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T \mathbb{E} \left[ \epsilon^2\right] \mathbf{X} \left\{ \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \right\}^T \\
    &= \sigma^2 \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \mathbf{X}^T  \mathbf{X} \left\{ \left( \mathbf{X}^T \mathbf{X} + \lambda \mathbf{I}_{\rm pp} \right)^{-1} \right\}^T \\
\end{align*}
